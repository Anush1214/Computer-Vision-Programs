```python
#image enhancement using contrast adjustment and histogram equalization using OpenCV
```

```python
import cv2
import matplotlib.pyplot as plt 
```

```python
img=cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
contrast=cv2.convertScaleAbs(img,alpha=1.5,beta=0)
histogram=cv2.equalizeHist(img)
```

```python
cv2.imshow("OG",img)
cv2.imshow("contrast",contrast)
cv2.imshow("histogram",histogram)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(2,2,1)
plt.hist(img.ravel(),256,range=[0,256])
plt.title("Og Image Histogram")

plt.subplot(2,2,2)
plt.hist(contrast.ravel(),256,range=[0,256])
plt.title("Contrast adjustedImage Histogram")

plt.subplot(2,2,3)
plt.hist(histogram.ravel(),256,range=[0,256])
plt.title("Equalized Image Histogram")
```
