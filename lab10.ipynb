```python
# Write a program to demonstrate Haar feature-based cascade classifiers for Face and Eye Detection on images.
```

```python
import cv2
```

```python
face_case=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
eyes_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_eye.xml")
```

```python
img=cv2.imread("image.jpg")
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces=face_case.detectMultiScale(grey,scaleFactor=1.1,minNeighbors = 5, minSize = (30, 30))
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    grey_roi=grey[y:y+h,x:x+w]
    color_roi = img[y:y+h, x:x+w]
    eyes = eyes_cascade.detectMultiScale(grey_roi)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(color_roi,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```

```python
cv2.imshow("face", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
