```python
#image filtering mechanisms using OpenCV
```

```python
import cv2
```

```python
img=cv2.imread("image.jpg")
```

```python
mean=cv2.blur(img,(3,3))
median=cv2.medianBlur(img,3)
laplacian=cv2.convertScaleAbs(cv2.Laplacian(img,cv2.CV_64F))
```

```python
cv2.imshow("Original",img)
cv2.imshow("Mean Filtered",mean)
cv2.imshow("Median Filetred",median)
cv2.imshow("Laplacian Filtered ",laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
