```python
#Implement Edge detection and Contour detection on images using openCV
```

```python
import cv2
```

```python
img=cv2.imread("image.jpg")
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur=cv2.GaussianBlur(gray,(3,3),0)
wide=cv2.Canny(blur,10,100)
mid=cv2.Canny(blur,30,150)
tight=cv2.Canny(blur,240,250)
```

```python
cv2.imshow("Original Image", img)
cv2.imshow("Wide Edged Image", wide)
cv2.imshow("Mid Edged Image", mid)
cv2.imshow("Tight Edged Image", tight)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
dilated=cv2.dilate(mid,kernel,iterations=1)
contours,_=cv2.findContours(dilated,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
image_copy=img.copy()
```

```python
cv2.drawContours(img,contours,-1,(0,255,0),2)
cv2.imshow("contours",image_copy)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
