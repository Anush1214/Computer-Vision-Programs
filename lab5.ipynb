```python
#Draw different geometric shapes and to write text on images using OpenCV.
```

```python
import numpy as np 
import cv2
```

```python

img=np.zeros((500,700,3),dtype=np.uint8)
```

```python
import cv2
cv2.line(img,(0,0),(50,50),(255,255,255),4)
cv2.rectangle(img,(50,50),(100,100),(255,255,255),5)
cv2.circle(img, (250, 250), 100, (0, 0, 255), 5)
cv2.arrowedLine(img, (100, 100), (400, 400), (0, 255, 0), 5)
```

```python
center=200,50
axes=100,30
angle=15
```

```python
cv2.ellipse(img,center,axes,angle,0,360,(255,0,0),2)
font=cv2.FONT_HERSHEY_PLAIN
cv2.putText(img,'OpenCV',(350,350),font,4,(0,255,0),2,cv2.LINE_AA)
```

```python
cv2.imshow("shapes",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
